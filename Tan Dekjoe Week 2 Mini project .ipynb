```python
import numpy as np
import matplotlib.pyplot as plt
#Experiment with circle
# Set the radius and center of the circle
radius = 5
center = (0, 0)

# Create a figure and axis
fig, ax = plt.subplots()


# Create a circle patch
circle = plt.Circle(center, radius, fill=False, edgecolor='black')

# Add the circle patch to the axis
ax.add_patch(circle)

# Set the limits of the plot
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Set aspect ratio to 'equal' to ensure the circle is not distorted
ax.set_aspect('equal')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Question 4 answer 
# Create a black image
gray = np.zeros((200, 200), dtype="uint8")

# Set the radius and center of the circle
radius = 50
center = (100, 100)

# Create a circle patch , plt.circle creates circle and sets it to center of black image , circle radius of radius = 50 , fill it up with color , and fill with color white.
circle = plt.Circle(center, radius, fill=True, color='white')

# Create a figure and axis
fig, ax = plt.subplots()

# Add the circle patch to the axis
ax.add_patch(circle)

# Display the black image
plt.imshow(gray, cmap=plt.cm.gray)
plt.title("Black Image with white circle")
plt.xticks([]), plt.yticks([])

# Display the plot
plt.show()
```
